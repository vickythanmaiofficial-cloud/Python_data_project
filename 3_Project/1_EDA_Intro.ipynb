```python
import ast
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv("D:\\Python_data_project\\1_basics\\data_jobs.csv")
```

```python
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x:ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US=df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='United States')]
```

```python
df_plot=df_DA_US['job_location'].value_counts().head(10).to_frame()
```

```python
df_plot
```

```python
sns.set_theme(style='ticks')
sns.barplot(data=df_plot,x='count',y='job_location',hue='count',palette='dark:b_r',legend=False)
sns.despine()
plt.title('Job Count In US')
plt.ylabel('')
plt.xlabel('No of Job Openings')
plt.show()
```

```python
fig, ax=plt.subplots(1,3)
dict_col={'job_work_from_home':'Work from Home','job_no_degree_mention':'Degree Required','job_health_insurance':'Health Insurance Required'}
for i,(key,value) in enumerate(dict_col.items()):
    df_DA_US[key].value_counts().plot(kind='pie',ax=ax[i],startangle=90,autopct='%1.1f%%')
    ax[i].set_title(value)
    ax[i].set_ylabel('')
fig.tight_layout()
```

```python
df_plot=df_DA_US['company_name'].value_counts().head(10).to_frame()
sns.set_theme(style='ticks')
sns.barplot(data=df_plot,x='count',y='company_name',hue='count',palette='dark:b_r',legend=False)
sns.despine()
plt.title('Company Openings In US')
plt.ylabel('')
plt.xlabel('No of Job Openings')
plt.show()
```
